Binarização

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Para exibir imagens no Colab
```

```python
# Carrega a imagem
img = cv2.imread('/content/Image/img_lena_2.jpg')

# Verifica se a imagem foi carregada corretamente
if img is None:
    print("Erro: Não foi possível carregar a imagem. Verifique o caminho.")
else:
    # Converte para escala de cinza
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detecta bordas usando o algoritmo Canny
    bordas = cv2.Canny(img_gray, 100, 200)

    # Combina as duas imagens lado a lado para exibição
    resultado = np.hstack([img_gray, bordas])

    # Exibe a imagem no Colab
    cv2_imshow(resultado)
```
